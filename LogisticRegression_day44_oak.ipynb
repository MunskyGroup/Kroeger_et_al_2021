```python
# import packages for uploading data
import re
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# update plot parameters
params = {'legend.fontsize': 16,
          'figure.figsize': (10, 10),
         'axes.labelsize': 20,
         'axes.titlesize':24,
         'xtick.labelsize':22,
         'ytick.labelsize':22}
plt.rcParams.update(params)
plt.style.use('grayscale')

from scipy.stats import ttest_ind as tt
from scipy.stats import linregress, norm
from sklearn.preprocessing import StandardScaler

# import k-fold function
from sklearn.model_selection import KFold

# import RF classifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc, confusion_matrix

from tqdm import tqdm

np.random.seed(123)
```

```python
table = pd.read_csv('data/day_44_oak_features.csv')

table.dropna(inplace=True)

sample_names = table['Sample ID'].values  
table.head()
```

```python
# set up X to be the matrix of features 
feature_names = ['Total Biomass', 'Bact Richness', 'Fungi Richness']
X = np.array(table[feature_names].values, np.double)
data_df = table[['Sample ID']+feature_names]

# set of DOC binary vector 
DOC = table['DOC'].values
DOC_cont = table['DOC_cont'].values
DOC_ctg = np.zeros(len(DOC))

for i, doc in enumerate(DOC):
    if doc == 'High':
        DOC_ctg[i] = 1
    else:
        DOC_ctg[i] = 0
        
sample2target = {s:t for (s,t) in zip(sample_names, DOC_ctg)}
```

```python
# eg. 100A --> 100 
def sample_parser(sample_id):
    return re.split("[ABCDEF]", sample_id)[0]

# function to create a dictionary mapping sample to a set of replicates
def get_sample_dict(data_df, sample_parser):
    # get the set of samples
    sample_ids = data_df.iloc[:, 0].values
    sample_ids_norep = [sample_parser(s) for s in sample_ids]
    # determine unique set of samples
    unique_samples = np.unique(sample_ids_norep)
    # create sample_dict to map sample id without rep to set of replicate samples 
    sample_dict = {}
    for s in unique_samples:
        # get all replicates of sample s
        sample_inds = np.in1d(sample_ids_norep, s)
        replicates = data_df.iloc[sample_inds, :]
        # store replicates in dict 
        sample_dict[s] = replicates
    # create a dataframe that has the sample names
    sample_df = pd.DataFrame()
    sample_df["Samples"] = unique_samples
    
    return sample_dict, sample_df

# function to return dataframes 
def get_Xtrain_Xtest(sample_dict, train_samples, test_samples):
    
    # create train dataframe
    for i, train_sample in enumerate(train_samples):
        # pull the dataframe for the sample
        if i==0:
            Xtrain_df = sample_dict[train_sample]
        else:
            Xtrain_df = pd.concat((Xtrain_df, sample_dict[train_sample]))
    
    # create test dataframe 
    for i, test_sample in enumerate(test_samples):
        # pull the dataframe for the sample
        if i==0:
            Xtest_df = sample_dict[test_sample]
        else:
            Xtest_df = pd.concat((Xtest_df, sample_dict[test_sample]))
    
    return Xtrain_df, Xtest_df
```

```python
# keep track of predictions
oof = []
ytrue = []

# create a dictionary mapping sample to a set of replicates
# k-folds will divide samples, keeping replicates contained within train and test folds
sample_dict, sample_df = get_sample_dict(data_df, sample_parser)
kf = KFold(n_splits=10)
kf.get_n_splits(sample_df)
for j,(train_index, test_index) in enumerate(kf.split(sample_df)):

    train_samples = sample_df.iloc[train_index].values.flatten()
    test_samples  = sample_df.iloc[test_index].values.flatten()
    Xtrain_df, Xtest_df = get_Xtrain_Xtest(sample_dict, train_samples, test_samples) 
    Xtrain, Xtest = np.array(Xtrain_df.values[:,1:], np.float), np.array(Xtest_df.values[:,1:], np.float)
    train_samples = Xtrain_df.iloc[:, 0].values
    test_samples = Xtest_df.iloc[:, 0].values
    
    ytrain = np.array([sample2target[s] for s in train_samples], np.float)
    ytest  = np.array([sample2target[s] for s in test_samples], np.float) 
    
    Xscaler = StandardScaler().fit(Xtrain)
    Xtrain = Xscaler.transform(Xtrain)
    Xtest  = Xscaler.transform(Xtest)
    
    LR_model = LogisticRegression()
    LR_model.fit(X=Xtrain, y=ytrain)
    preds = LR_model.predict_proba(Xtest)

    oof.append(preds)
    ytrue.append(ytest.ravel())
    
ytrue = np.concatenate(ytrue)
ypred = np.concatenate(oof)
ypredproba = np.array([y[1] for y in ypred])
ypredmax   = np.round(ypredproba)
```

```python
fpr, tpr, threshold = roc_curve(np.array(ytrue, np.int), ypredproba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(9, 6))
plt.title('Oak Receiver Operating Characteristic')
plt.plot(fpr, tpr, label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'k--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.savefig("results/logistic_regression/Logistic_Regression_ROC_Curve_Oak.png", dpi=150)
plt.show()
```

```python
# import stats library
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.discrete.discrete_model import Logit
import statsmodels.api as sm 
```

```python
# standardize X and add a constant for LR model 
X_std = StandardScaler().fit(X).transform(X)
X_L = sm.add_constant(X_std)

# fit LR model to data 
L = Logit(DOC_ctg, X_L)
results = L.fit()

coef = results.params
std_err = np.diagonal(results.cov_params())**.5
wald = (coef / std_err)**2
pvalues = results.pvalues

results.summary()
```

```python
import scipy

def Z_proportions(p1, p2, n1, n2):
    P = (p1*n1 + p2*n2) / (n1 + n2)
    Z = (p1 - p2) / np.sqrt(P*(1-P) * (1/n1 + 1/n2))
    return Z

baseline = np.max([len(ytrue)-np.sum(ytrue), np.sum(ytrue)]) / len(ytrue)
acc = np.sum(ypredmax==ytrue) / len(ytrue)
z_score = Z_proportions(acc, baseline, len(ypredmax), len(ytrue))
model_p_value = scipy.stats.norm.sf(z_score)
print("p value on out-of-fold predictions: {}".format(model_p_value))
```

```python
fs_df = pd.DataFrame()
fs_df['Factor'] = feature_names
fs_df['Coefficient'] = coef[1:]
fs_df['Wald stat.'] = wald[1:]
fs_df['P values'] = pvalues[1:]
fs_df["Model P value"] = [model_p_value] + [""]*(len(feature_names)-1)

fs_df.to_csv('results/logistic_regression/FeatureImportances_oak.csv', index=False)
fs_df.head(10)
```
